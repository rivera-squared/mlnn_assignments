### Question 1-- Selecting Wednesday ['2'] as n

```python
# max_depth = 4
```

```python
hours = bikeshare.index.values.reshape(-1,1)

bike_reg = tree.DecisionTreeRegressor(max_depth=4)
bike_reg.fit(hours, bikeshare['2'].fillna(0))

bike_pred = bike_reg.predict(hours)
wed = bikeshare['2'].fillna(0)

plt.scatter(hours, bike_pred, s=2)
plt.scatter(hours, wed, s=2)
```

```python
mean_squared_error(wed, bike_pred, squared = False) # y_true, y_pred
```

Based on the groups of MSE values derived from the models, I would recommend a max_depth of 8. First, controlling for the day of the week, the model's mean squared error kept decreasing, a good indication that the model is being improved as the max_depth increased. Since overfitting is a concern for deploying the model for future datasets, I calculated the percentage change of MSE with an increased max depth from the previous max depth value selected--sort of like a diminishing returns. To ensure that max_depth = 8 did not overfit the model, I continued calculating the MSE and the diminishing returns for depths of 9, 10, etc... The MSEs continued to decreased as expected and the plots of actual and predicted values began to converge into what it seemed as a single plot, a good confirmation that after max = 8, the model was overfitting. In addition, the diminishing returns from the increased max_depths beyond 8 began to decrease significantly; in other words, it was not paying off increasing the max_depth beyond 8. 

A similar scenario occured when the model was being controlled by the max_depth instead of the day of the week. Among the different days of the week, the max_depth = 8, was consistent througout. The MSEs were the lowest and the scenario of the diminishing returns beyond max_depth =8 also increased.
